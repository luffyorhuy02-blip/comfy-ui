```python
#@title 1Ô∏è‚É£ KEEP ALIVE (Ch·∫°y ng·∫ßm)
import time, random, threading, gc
from IPython.display import display, Javascript

# JavaScript Keep Alive
display(Javascript('''
(function(){
    let count = 0;
    function keepAlive() {
        count++;
        console.log("Keep alive tick: " + count);
        document.querySelector("colab-connect-button")?.click();
        
        // Sim UI update
        var status = document.getElementById("ka-status");
        if (status) status.innerHTML = "üîÑ Active: " + count;

        setTimeout(keepAlive, 25000 + Math.random() * 15000);
    }
    
    // Create Status UI
    var div = document.createElement("div");
    div.id = "ka-status";
    div.innerHTML = "üîÑ ON";
    div.style.cssText = "position: fixed; top: 10px; right: 10px; background: linear-gradient(135deg, #FF6B6B, #556270); color: white; padding: 8px 15px; border-radius: 20px; z-index: 9999; font-family: sans-serif; box-shadow: 0 4px 6px rgba(0,0,0,0.1);";
    document.body.appendChild(div);
    
    keepAlive();
})();
'''))

# Python Keep Alive Background Thread
def py_keep_alive():
    while True:
        time.sleep(random.randint(30, 90))
        gc.collect() # Memory cleanup to fake activity
        
threading.Thread(target=py_keep_alive, daemon=True).start()
print("‚úÖ Keep Alive activated!")
```

```python
#@title üîü üöÄ CH·∫†Y COMFYUI
import subprocess
import threading
import time
import socket
from google.colab import userdata

# --- C·∫§U H√åNH ---
try:
    TOKEN = userdata.get('CLOUDFLARE_TOKEN')
except:
    TOKEN = ""
    print("‚ö†Ô∏è Ch∆∞a c·∫•u h√¨nh Secrets! Vui l√≤ng th√™m 'CLOUDFLARE_TOKEN' v√†o Colab Secrets.")
    print("üëâ Ho·∫∑c nh·∫≠p tr·ª±c ti·∫øp v√†o code n·∫øu ƒëang test (kh√¥ng truy√™n git).")

DOMAIN = "comfyui.maymoc.shop"
# ----------------

def start_tunnel():
    if not TOKEN:
        print("‚ùå Kh√¥ng t√¨m th·∫•y Token! Tunnel s·∫Ω kh√¥ng ch·∫°y.")
        return
    
    print(f"üöÄ ƒêang kh·ªüi ƒë·ªông Tunnel t·ªõi {DOMAIN}...")
    cmd = ["cloudflared", "tunnel", "run", "--token", TOKEN]
    subprocess.run(cmd)

# Ch·∫°y tunnel trong thread ri√™ng
threading.Thread(target=start_tunnel, daemon=True).start()

print("‚è≥ ƒêang ƒë·ª£i ComfyUI kh·ªüi ƒë·ªông... (Vui l√≤ng ch·ªù 1-2 ph√∫t)")
time.sleep(10)

print("="*50)
print(f"üåç Truy c·∫≠p ComfyUI t·∫°i: https://{DOMAIN}")
print("="*50)

%cd /content/ComfyUI
!python main.py --listen 0.0.0.0 --port 8188
```
